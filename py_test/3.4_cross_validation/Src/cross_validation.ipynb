```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt 

#加载数据
iris = datasets.load_iris()
X1 = iris.data
y1 = iris.target

wine = datasets.load_wine()
X2 = wine.data
y2 = wine.target

#数据可视化
plt.scatter(X1[:,0],X1[:,1],c=y1)
plt.show()

plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import LeaveOneOut

#加载模型
log_model = LogisticRegression()

#采用10折交叉验证
y_pred = cross_val_predict(log_model, X1, y1, cv=10)
print("Accuracy:",metrics.accuracy_score(y1, y_pred))

#采用留一法交叉验证
loo = LeaveOneOut()
accuracy = 0
# 每次留一个样本作为测试集，其他作为训练集
for train, test in loo.split(X1):
    # 在训练集上训练模型
    log_model.fit(X1[train], y1[train]) 
    # 在测试集上预测模型
    y_p = log_model.predict(X1[test])
    if y_p == y1[test] : accuracy += 1  
print(accuracy / np.shape(X1)[0])


```

```python
#加载模型
log_model = LogisticRegression()

#采用10折交叉验证
y_pred = cross_val_predict(log_model, X2, y2, cv=10)
print("Accuracy:",metrics.accuracy_score(y2, y_pred))

#采用留一法交叉验证
loo = LeaveOneOut()
accuracy = 0
for train, test in loo.split(X2):
    # 在训练集上训练模型
    log_model.fit(X2[train], y2[train]) 
    # 在测试集上预测模型
    y_p = log_model.predict(X2[test])
    if y_p == y2[test] : accuracy += 1  
print(accuracy / np.shape(X2)[0])
```
